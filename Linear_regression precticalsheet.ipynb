```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
weeks_data = [1,2,3,4,5]
sales_data = [2,4,5,4,5]

df = pd.DataFrame({
    "Weeks":weeks_data,
    "Sales":sales_data
})
df
```

```python
X = df["Weeks"].values
Y = df["Sales"].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

n = len(X)   
numer = 0 
denom = 0 

for i in range(n):
    numer+= (X[i] - mean_y) * (Y[i] - mean_y)
    denom+= (X[i] - mean_x) ** 2
m = numer/denom
c = mean_y - (m * mean_x)

print(f"M : {m}")
print(f"C : {c}")
```

```python
arr_new =[]
for i in range(len(X)):
    y = m * X[i] + c
    arr_new.append(y)
```

```python
x1 = arr_new
print(x1)
```

```python

plt.plot(X,x1,color='red',label='Regressin Line',linestyle='--')
plt.scatter(X,Y,color="blue",label='Data Points',s=48)
plt.xlabel("Weeks")
plt.ylabel("Sales")
plt.title("Linear Regression  Demo")
plt.grid(True,alpha=1,linestyle=':')
plt.legend(loc='upper left')
plt.show()
```

```python
max_x = np.max(X) + 1
min_x = np.min(X) - 1

x= np.linspace(min_x,max_x) 

y = (m * x) + c

plt.plot(x,y,color='red',label='Regressin Line',linestyle='--')
plt.scatter(X,Y,color="blue",label='Data Points',s=48)
plt.xlabel("Weeks")
plt.ylabel("Sales")
plt.title("Linear Regression  Demo")
plt.grid(True,alpha=1,linestyle=':')
plt.legend(loc='upper left')
plt.show()
```

```python
##r2
ss_tot = 0  
ss_res = 0 

for i in range(n):
    y_pred = m * X[i] + c
    ss_tot +=(Y[i] - mean_y) ** 2 
    ss_res +=(Y[i] - y_pred)
r2 = 1 - (ss_res/ss_tot) 
print(f"r2 : {r2}")
```
